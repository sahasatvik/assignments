```python
from matplotlib.pyplot import *
from numpy import *
```

```python
def plot_f(f, limits, xlo, xhi, ylo, yhi):
    a, b = limits
    g = lambda x: f((x - a) % (b - a) + a)
    x = linspace(xlo, xhi, 1000)
    y = linspace(ylo, yhi, 1000)
    fig = figure()
    ax = fig.add_subplot(1, 1, 1)

    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    gca().set_aspect('equal', adjustable='box')
    plot(x, g(x))
    axis((xlo, xhi, ylo, yhi))
```

```python
rcParams["figure.figsize"]= 10, 6
rcParams["figure.dpi"]= 100

plot_f(lambda x: x, (-pi, +pi), -12, 12, -pi, +pi)
```

```python
plot_f(lambda x: x, (0, 2*pi), -12, 12, 0, 2*pi)
```

```python
plot_f(lambda x: sin(pi * x), (-1.0/2, 1.0/2), -2, 2, -1.2, 1.2)
```

```python
plot_f(lambda x: sin(pi * x), (0, 1), -2, 2, 0, 1.2)
```

```python
plot_f(lambda x: where(x < 0, x + x**2, x - x**2), (-1, 1), -4, 4, -0.3, 0.3)
```

```python
plot_f(lambda x: where(x < 0, 0, 1), (-1, 3), -10, 10, 0, 1.2)
```

```python
plot_f(lambda x: where(x < 2, x/2.0, 1), (0, 3), -10, 10, 0, 1.2)
```
