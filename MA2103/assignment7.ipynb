```python
from matplotlib.pyplot import *
from numpy import *
```

```python
def f_p(f, a=-1, b=1):
    return lambda x: f((x - a) % (b - a) + a)

f = lambda x: where(x < 0, 3*x**2 + 2*x**3, 3*x**2 - 2*x**3)
f1 = lambda x: where(x < 0, 6*(x + x**2), 6*(x - x**2))
f2 = lambda x: where(x < 0, 6*(1 + 2*x), 6*(1 - 2*x))
f3 = lambda x: where(x < 0, 12, -12)

def ff(x, n):
    return 0.5 - np.sum(48 * cos((2*i + 1) * pi * x) / ((2*i + 1) * pi)**4 for i in range(n))
def ff1(x, n):
    return np.sum(48 * sin((2*i + 1) * pi * x) / ((2*i + 1) * pi)**3 for i in range(n))
def ff2(x, n):
    return np.sum(48 * cos((2*i + 1) * pi * x) / ((2*i + 1) * pi)**2 for i in range(n))
def ff3(x, n):
    return -np.sum(48 * sin((2*i + 1) * pi * x) / ((2*i + 1) * pi) for i in range(n))
```

```python
xlo, xhi = -2, 2

x = linspace(xlo, xhi, 1000)
fig = figure()

rcParams["figure.figsize"]= 10, 6
rcParams["figure.dpi"]= 100

plot(x, f_p(f)(x), 'yellow', label='$f(x)$')
plot(x, ff(x, 1), '--r', label='1 term')
legend()
```

```python
rcParams["figure.figsize"]= 10, 6
rcParams["figure.dpi"]= 100

plot(x, f_p(f1)(x), 'yellow', label="$f'(x)$")
plot(x, ff1(x, 1), ':r', label='1 term')
plot(x, ff1(x, 2), '--k', label='2 terms')
legend()
```

```python
rcParams["figure.figsize"]= 10, 6
rcParams["figure.dpi"]= 100

plot(x, f_p(f2)(x), 'yellow', label="$f''(x)$")
plot(x, ff2(x, 1), ':r', label='1 term')
plot(x, ff2(x, 2), ':g', label='2 terms')
plot(x, ff2(x, 3), ':m', label='3 terms')
plot(x, ff2(x, 5), '--k', label='5 terms')
legend()
```

```python
rcParams["figure.figsize"]= 10, 6
rcParams["figure.dpi"]= 100

plot(x, f_p(f3)(x), 'yellow', label="$f'''(x)$")
plot(x, ff3(x, 1), ':r', label='1 term')
plot(x, ff3(x, 3), ':g', label='3 terms')
plot(x, ff3(x, 7), ':m', label='5 terms')
plot(x, ff3(x, 15), '--k', label='11 terms')
plot(x, ff3(x, 51), ':b', label='51 terms')
legend()
```
